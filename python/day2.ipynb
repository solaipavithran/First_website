```python
def calculate():
    num = 1
    def inner_func():
        nonlocal num
        num += 2
        return num
    return inner_func
odd = calculate()

print(odd())
print(odd())
```

```python
def decor1(func):
    def inner():
        x = func()
        return x*x
    return inner
def decor(func):
    def inner():
        x =func()
        return 2*x
    return inner
@decor1
@decor
def num():
    return 10

@decor
@decor1
def num2():
    return 10
print(num())
print(num2())
```

```python
def make_timer():
    count = 0
    def timer(func):
        def wrapper():
            nonlocal count
            count+=1
            print(f"Function {func.__name__} called {count}")
            return func()
        return wrapper
    return timer
timer = make_timer()

@timer
def greet():
    print("hello")
greet()
greet()
```

```python
'''Create a function greet that
takes a name as positional arg
accepts greeeting using *args.
accepts a punctuation keyword-only argu (default  is "!").
test the function with muntiple greeting and different punctuation style.
'''

def greet(name,/,*arg,punctuation = "!"):
    s = ""
    for i in arg:
        s += i + " "
    print(f"{name} {s}{punctuation}")
greet("pavithran","good morning","welcome")
greet("vicky","good morning","welcome",punctuation= ".")
```

```python
def order_summery(item_price,quantity,discount = 0,tax = 5):

    p = item_price * quantity
    dis = p*(discount/100)
    tax_p = (p-dis)*(tax/100)
    total = (p-dis) + tax_p
    return total 
print(order_summery(100,1,5,10))
```
