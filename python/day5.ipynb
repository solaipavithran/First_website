```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
 
class Employee(Person):
    def __init__(self, name,age, emp_id):
        super().__init__(name, age)
        self.emp_id = emp_id
 
emp = Employee("John Doe",30,12345)
print(emp.name)
print(emp.age)
print(emp.emp_id)
 
```

```python
from multipledispatch import dispatch

@dispatch(int,int)
def product(first,sec):
    result = first*sec
    print(result)
@dispatch(int,int,float)
def product(first,sec,third):
    result = first + sec +third
    print(result)
@dispatch(float,float,int)
def product(first,sec,third):
    result = first * sec * third
    print(result)


product(2,3)
product(2,3,9.5)
product(2.1,3.5,6)
```

```python
class number: 
    def __init__(self, f, s): 
        self.f = f 
        self.s = s 
    def __iter__(self): 
        self.current = self.f 
        return self 
    def __next__(self): 
        if self.current <= self.s: 
            value = self.current 
            self.current += 1 
            return value 
        else: 
            raise StopIteration 
result = number(1, 5) 
for num in result:
    print(num)
```

```python
class circle:
    def __init__(self,r):
        self._radius = r
    def get_radius(self):
        return self._radius
    def set_radius(self,value):
        if value<0:
            raise ValueError("radiuscannot be -ve")
        self._radius = value
    def del_radius(self):
        print("delecting Radius")
        del self._radius
    c = circle()
    print(c._radius)
```

```python
#property
 
class Circle:
    def __init__(self,radius):
        self._radius = radius #protected attribute
 
        #getter method
    def get_radius(self):
        return self._radius
    # setter method
 
    def set_radius(self,value):
        if value<0:
            raise ValueError("radius cannot be negative")
        self._radius = value
 
    #delete method
    def del_radius(self):
        print("deleting radius")
        del self._radius
 
    radius = property(get_radius, set_radius, del_radius)
```
