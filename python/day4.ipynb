```python
# file can be open in excel
import csv
fields = ['name','branch','year','cgpa']
rows = [['rui','it','2','9'],['pavithran','it','4','8'],['vicky','it','2','9']]

with open('test.csv','w',newline="") as csvf:
    csvw = csv.writer(csvf,dialect ='excel')
    csvw.writerow(fields)
    csvw.writerows(rows)
```

```python
import re
pattern = r'www\.'
pattern2 = r'#[a-z]'
l = []
l1 = []
with open('new.txt','r') as f:
    for i in f.read().split(" "):
        if re.match(pattern,i):
            l.append(i)
        if re.match(pattern2,i):
            l1.append(i)
print(l,l1)
```

```python
# custom error

class NegativeNumber(Exception):
    pass
try:
    num = int(input("enter a positive number:"))
    if num<0:
        raise NegativeNumber("not allowed")
    print(f'you enter :{num}')
except NegativeNumber as e:
    print(e)
```

```python
# chained exceptions
try:
    try:
        num = int(input("Enter a num:"))
        if num == 0:
            raise ValueError("Zero is not allowed")
    except ValueError as ve:
        print(f"Inner Exception:{ve}")
        raise KeyError("An additional key error ")
except KeyError as ke:
    print(f"Outer Exception:{ke}")
```
